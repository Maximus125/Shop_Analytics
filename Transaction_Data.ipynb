```python
import re
name = list(df['PROD_NAME'])                 #Extracting weight from product
wt = []
for n in name:
    wt.append(re.sub("\D", "", n))
    
df['PROD_WT'] = wt
df['PROD_WT'] = df['PROD_WT'].astype('int64')
```
